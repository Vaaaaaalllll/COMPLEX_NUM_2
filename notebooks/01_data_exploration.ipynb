# AIS Data Exploration

This notebook demonstrates how to explore AIS data before and after cleaning.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

from src.cleaner import AISCleaner
from src.config import load_config

sns.set_style('whitegrid')
%matplotlib inline
```

## Load Data

```python
# Load raw data
input_file = Path('../data/sample/sample_ais.parquet')
df_raw = pd.read_parquet(input_file)
print(f'Raw data shape: {df_raw.shape}')
df_raw.head()
```

## Data Quality Analysis

```python
# Check for missing values
df_raw.isnull().sum()
```

```python
# Basic statistics
df_raw.describe()
```

## Clean Data

```python
# Initialize cleaner
config = load_config()
cleaner = AISCleaner(config)

# Clean data
df_clean = cleaner.clean_dataframe(df_raw)
print(f'Clean data shape: {df_clean.shape}')
print(f'Removed {len(df_raw) - len(df_clean)} records')
```

## Compare Before and After

```python
# Plot speed distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].hist(df_raw['speed'], bins=50, alpha=0.7)
axes[0].set_title('Speed Distribution (Raw)')
axes[0].set_xlabel('Speed (knots)')

axes[1].hist(df_clean['speed'], bins=50, alpha=0.7, color='green')
axes[1].set_title('Speed Distribution (Clean)')
axes[1].set_xlabel('Speed (knots)')

plt.tight_layout()
plt.show()
```
